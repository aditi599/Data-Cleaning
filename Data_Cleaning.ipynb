Started the cleaning process column by column, in this case, started with the column square feet

Here the data is being replaced with the blank where the suffix is 'sqft' , 'sqm' , 'sqyrd' ,'rood','acre','Move','Ready','ground','To','Resale','to' , or we can use use str.contains method to replace all the characters that are within the range of A-z

```python
distinct_sqft=df['square_feet'].unique()
df['square_feet']=df['square_feet'].str.replace('sqft','')
df['square_feet']=df['square_feet'].str.replace('sqm','')
df['square_feet']=df['square_feet'].str.replace('sqyrd','')
df['square_feet']=df['square_feet'].str.replace('rood','')
df['square_feet']=df['square_feet'].str.replace('acre','')
df['square_feet']=df['square_feet'].str.replace('Move','')
df['square_feet']=df['square_feet'].str.replace('Ready','')
df['square_feet']=df['square_feet'].str.replace('ground','')
df['square_feet']=df['square_feet'].str.replace('To','')
df['square_feet']=df['square_feet'].str.replace('Resale','')
df['square_feet']=df['square_feet'].str.replace('to','')
df['square_feet']=df['square_feet'].str.replace('sqft','',regex=False)
ab=df['square_feet'].str.contains('qwertyuioplkjhgf,dsazxcvbnm')
ab.nunique()
df['square_feet']
df
```
